```python
# Inialize SQL Connection
import pymysql
# Connect to XAMPP MySQL
sp = pymysql.connect(
    host='localhost',
    user='root',
    password='',         
)

cursor = sp.cursor()
print("PyMySQL connection established!")
```

```python
# creating securecheck database 
cursor.execute("CREATE DATABASE SECURECHECK")
```

```python
# use database
cursor.execute("USE SECURECHECK")
sp.commit()
```

```python
# Table creation
cursor.execute("""CREATE TABLE SECURECHECK.logs (
id INT AUTO_INCREMENT PRIMARY KEY,
stop_date DATE,
stop_time TIME,
country_name VARCHAR(50),
driver_gender VARCHAR(10),
driver_age INT,
driver_race VARCHAR(50),
violation_raw VARCHAR(100),
violation VARCHAR(100),
search_conducted BOOLEAN,
search_type VARCHAR(100),
stop_outcome VARCHAR(100),
is_arrested BOOLEAN,
stop_duration VARCHAR(50),
drugs_related_stop BOOLEAN,
vehicle_number VARCHAR(50))
""")
```

```python
sp.commit()
print("Table 'logs' created successfully in pymysql!")
```

```python
df.info()
```

```python
#insert police ledger data into SQL Table
for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO SECURECHECK.logs (
            stop_date, stop_time, country_name, driver_gender, driver_age,
            driver_race, violation_raw, violation, search_conducted, search_type,
            stop_outcome, is_arrested, stop_duration, drugs_related_stop, vehicle_number
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['stop_date'], row['stop_time'], row['country_name'], row['driver_gender'], row['driver_age'],
        row['driver_race'], row['violation_raw'], row['violation'], row['search_conducted'],
        row['search_type'], row['stop_outcome'], row['is_arrested'], row['stop_duration'],
        row['drugs_related_stop'], row['vehicle_number']
    ))

sp.commit() 
```

```python
cursor.close()
sp.close()
```
